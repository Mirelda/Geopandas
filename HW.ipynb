```python
import geopandas
import matplotlib.pyplot as plt
import pandas as pd 
from shapely.geometry import Polygon
wind = pd.read_excel("wind_farms.xlsx")
ilce = geopandas.read_file("Ilceler_region.shp")
```

```python
wind.iloc[:,]
# lAT = [K1_lAT, K2_LAT, U1_LAT,U2_LAT]
# LON = [K1_LON, K2_LON , U2_LON, U2_LON.1]
# data = [MWm(:,1),MWe(:,2),POLYGON]
# data = geopandas.GeoDataFrame()

```

```python
lat = []
lon = []
for i in range(163):
    x = list([wind["K1_LAT"].iloc[i],wind["K2_LAT"].iloc[i],wind["U1_LAT"].iloc[i],wind["U2_LAT"].iloc[i]])
    lat.append(x)
    y = list([wind["K1_LON"].iloc[i],wind["K2_LON"].iloc[i],wind["U2_LON"].iloc[i],wind["U2_LON.1"].iloc[i]])
    lon.append(y)

```

```python
wind_new = wind.iloc[:,1:3]
wind_new.head()
```

```python
All_Poly = []
for i in range(163):
    p = Polygon(zip(lon[i],lat[i]))
    All_Poly.append(p)
```

```python
gdf = geopandas.GeoDataFrame(wind_new, geometry = All_Poly, crs = 'EPSG:4326')
gdf.plot()
```

```python
wind_final = geopandas.sjoin(ilce,gdf, how = "inner",op = "contains")
wind_final.head(20)
```

```python
wind_final.plot(cmap = "rainbow")
```

```python
wind_final.to_file(driver = "ESRI Shapefile",filename = "wind_farms.shp", encoding = "utf-8")
```
